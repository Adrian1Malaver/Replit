##GRAFICA POR CADA CLIENTE SEGUN LA CANTIDAD DE PRODUCTOS COMPRADOS POR CADA NOMBRE DEL PRODUCTO:

se utilizo el bucle for para iterar a traves de cada cliente dentro del bucle, se filtran las filas correspondientes a las facturas y productos asociados a ese cliente en particular. El bucle itera sobre cada ID_Cliente y dentro de el se realiza todo el analisis y la cracion del grafico especifico para ese cliente. Esto se repite para cada cliente, generando visualizaciones individuales y mensajes de preferencia de compra para cada uno.


```python
import pandas as pd
import matplotlib.pyplot as plt

#Cargar los datos de los archivos CSV
factura_df = pd.read_csv('FACTURA.csv')
clientes_df = pd.read_csv('CLIENTES.csv')

#Número total de clientes
num_clientes = clientes_df['ID_Cliente'].nunique()

for cliente_id in range(num_clientes):
    #filas correspondientes al ID_Cliente actual en CLIENTES.csv
    facturas_cliente = clientes_df[clientes_df['ID_Cliente'] == cliente_id]['ID_FACTURA']

    #filas correspondientes a las facturas del cliente actual en FACTURA.csv
    factura_cliente = factura_df[factura_df['ID_Factura'].isin(facturas_cliente)]

    #cantidad total por producto para el cliente actual
    cantidad_total_por_producto = factura_cliente.groupby('Producto')['Cantidad'].sum().reset_index()

    #producto más comprado y su cantidad para el cliente actual
    producto_mas_comprado = cantidad_total_por_producto.loc[cantidad_total_por_producto['Cantidad'].idxmax()]

    #gráfico de barras para el cliente actual
    plt.figure(figsize=(10, 6))
    plt.bar(cantidad_total_por_producto['Producto'], cantidad_total_por_producto['Cantidad'])
    plt.xlabel('Producto')
    plt.ylabel('Cantidad Comprada')
    plt.title(f'Cantidad Comprada por Producto para el Cliente {cliente_id}')

    #preferencia del cliente actual
    print(f"Preferencia del cliente {cliente_id}: {producto_mas_comprado['Producto']} con {int(producto_mas_comprado['Cantidad'])} unidades vendidas.")

    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.show()
```
